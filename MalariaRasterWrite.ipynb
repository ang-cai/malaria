```python
import rioxarray
import numpy
```

```python
lst = rioxarray.open_rasterio("StandardizedUganda/standard_uganda_LSTday_2km_2000.tif")
```

```python
rainfall = rioxarray.open_rasterio("StandardizedUganda/standard_uganda_Rainfall_CHIRPS_2km_2000.tif")
```

```python
lst.plot()
```

```python
rainfall.plot()
```

```python
malaria_logit_prevalence = (-1.5 + .5 * lst + .75 * rainfall)
```

```python
malaria_logit_prevalence.plot()
```

```python
malaria_prevalence_probability = 1 / (1 + numpy.exp(-malaria_logit_prevalence))
```

```python
malaria_prevalence_probability.plot()
```

```python
malaria_prevalence_probability.rio.to_raster("StandardizedUganda/mock_malaria_prevelance_uganda_2km_2000.tif")
```
