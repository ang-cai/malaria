# Turn tif into png

```python
import rasterio
import matplotlib.pyplot as plt
import matplotlib
```

```python
paths = ["Uganda Covariate Rasters/", "Uganda Standardized Rasters/"]
tiffs = ["uganda_Rainfall_CHIRPS_2km_2018", "uganda_elevation_2km", "uganda_LSTday_2km_2018"]
titles = ["Rainfall", "Elevation", "Land Surface Temperature"]
```

```python
figure, axis = plt.subplots()
raster = "Uganda Malaria Data/uganda_mock_malaria_prevalence_2km_2018.tif"
with rasterio.open(raster) as src:
    image = src.read(1)
image = axis.imshow(image, cmap="viridis")
axis.set_xlabel("x")
axis.set_ylabel("y")
figure.colorbar(image)
figure.suptitle("Malaria Prevalence = -1.5  + .75∙Rainfall + .25∙Elevation + .5∙Temperature ")
file = "Uganda Malaria Data/uganda_mock_malaria_prevalence_2km_2018.png"
plt.savefig(file, bbox_inches="tight", pad_inches=0) 
plt.show()
```

```python
for path in paths:
    for i in range(len(tiffs)):
        figure, axis = plt.subplots()
        if path == "Uganda Standardized Rasters/":
            raster = f"{path}standard_{tiffs[i]}.tif"
        else: 
            raster = f"{path}{tiffs[i]}.tif"

        with rasterio.open(raster) as src:
            image = src.read(1)
            extent = (src.bounds.left, src.bounds.right, src.bounds.bottom, 
                        src.bounds.top)
            
        if path == "Uganda Standardized Rasters/":
            image = axis.imshow(image, cmap="coolwarm", extent=extent, vmin=-10,
                                vmax=10)
        else:
            image = axis.imshow(image, cmap="viridis", extent=extent)

        axis.set_xlabel("x")
        axis.set_ylabel("y")
        figure.colorbar(image)
        figure.suptitle(titles[i])

        if path == "Uganda Standardized Rasters/":
            file = f"{path}standard_{tiffs[i]}.png"
        else: 
            file = f"{path}{tiffs[i]}.png"
        plt.savefig(file, bbox_inches="tight", pad_inches=0) 
        plt.show()
```

```python
for path in paths:
    for i in range(len(tiffs)):
        figure, axis = plt.subplots()
        if path == "Uganda Standardized Rasters/":
            raster = f"{path}standard_{tiffs[i]}.tif"
        else: 
            raster = f"{path}{tiffs[i]}.tif"

        with rasterio.open(raster) as src:
            image = src.read(1)
            extent = (src.bounds.left, src.bounds.right, src.bounds.bottom, 
                        src.bounds.top)
            
        if path == "Uganda Standardized Rasters/":
            image = axis.imshow(image, cmap="coolwarm", extent=extent, vmin=-10,
                                vmax=10)
            matplotlib.colors.Normalize(vmin=-10, vmax=10)
        else:
            image = axis.imshow(image, cmap="viridis", extent=extent)

        axis.set_xlabel("x")
        axis.set_ylabel("y")
        figure.colorbar(image)
        figure.suptitle(titles[i])

        if path == "Uganda Standardized Rasters/":
            file = f"{path}standard_{tiffs[i]}.png"
        else: 
            file = f"{path}{tiffs[i]}.png"
        plt.savefig(file, bbox_inches="tight", pad_inches=0) 
        plt.show()
```
